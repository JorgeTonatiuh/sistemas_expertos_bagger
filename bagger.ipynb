```python
class Bagger:

  def __init__(self):    
    # Dictionary of items, 
    # key: id number 
    # value: array of strings -> [item, container_type, size, isFrozen]
    self.objects = {            
        0: ["Bread", "Plastic bag", "Medium", "No"],
        1: ["Glop", "Jar", "Small", "No"],
        2: ["Granola", "Cardboard box", "Large", "No"],
        3: ["Ice cream", "Cardboard carton", "Medium", "Yes"],
        4: ["Pepsi", "Bottle", "Large", "No"],
        5: ["Chips", "Plastic bag", "Medium", "No"]     
    }

    # Initial step        
    self.step = "" 

    # Begins process
    self.empty_bags()
    self.show_market_items()
    self.add_items()
    self.verify_items()
    self.list_items_to_buy()
    self.perform_rules()
    self.show_total_bags()

  def show_total_bags(self):
    print("Packing finished.\n")
    self.total_bags_of("Large",self.large_bag,self.large_bag_bag)
    self.total_bags_of("Medium",self.med_bag,self.med_bag_bag)
    self.total_bags_of("Small",self.small_bag,self.small_bag_bag)

  def total_bags_of(self,type_bag,bag,bag_of_bags):
    print(type_bag," bags:")
    if len(bag) > 0:
      bag_of_bags.append(bag)
    for items in bag_of_bags:
      print("Bags:",items)

  def show_market_items(self):
    print("Market items: ")
    print("id\tItem")
    for key,values in self.objects.items():
      print(key,"\t",values[0])

  def add_items(self):
    print("\n")
    print("---------------------")
    print("Add the ID of the items to buy")
    self.items_id = input().split()
    self.items_id = [int(item) for item in self.items_id]

  def verify_items(self):
    print("---------------------")
    print("Checking items...")
    self.name_of_items = self.id_to_item(self.items_id)

  def list_items_to_buy(self):
    print("---------------------")
    print("Items to buy:")    
    if len(self.items_id) > 0:
      print("\t",self.name_of_items)
      self.step = "check-order"
    else:
      print("\tNo items to buy.")

  def perform_rules(self):
    print("---------------------")
    while(self.step != "final"):
      self.apply_rules()

  def apply_rules(self):
    large_list = [item for item in self.items_id if self.objects[item][2] == "Large"]
    med_list = [item for item in self.items_id if self.objects[item][2] == "Medium"]
    small_list = [item for item in self.items_id if self.objects[item][2] == "Small"]

    while(True):
      # B1
      if self.step == "check-order" and (5 in self.items_id) and (4 not in self.items_id):
        print("Â¿Do you want to add a Pepsi to your list of items? Y/N")
        ans = input()
        if ans == "Y":
          self.items_id.append(4)	
        print("Final list of items:")
        pedido_letra = self.id_to_item(self.items_id)
        print("\t",self.name_of_items)
        break

			#B2
      if self.step == "check-order":
        self.step = "pack-large-items"
        print("Packing large items...")
        break

			#B3
      if self.step == "pack-large-items" and len(large_list) > 0 and (4 in large_list) and len(self.large_bag) < 6:
        self.large_bag.append(self.objects[4][0])
        self.items_id.remove(4)
        break

			#B4
      if self.step == "pack-large-items" and len(large_list) > 0 and len(self.large_bag) < 6:
        item = large_list.pop()
        self.large_bag.append(self.objects[item][0])
        self.items_id.remove(item)
        break

			#B5
      if self.step == "pack-large-items" and len(large_list) > 0:
        self.get_new_bag()
        break

			#B6
      if self.step == "pack-large-items":
        self.step = "pack-medium-items"
        print("Packing medium items...")
        break

			#B7
      if self.step == "pack-medium-items" and len(med_list) > 0 and len(self.med_bag) < 6 and (3 in med_list) and (3 not in self.frozen_bag):
        self.frozen_bag.append(self.objects[3][0])
        self.med_bag.append(self.frozen_bag)
        self.frozen_bag = []
        self.items_id.remove(3)
        break

			#B8
      if self.step == "pack-medium-items" and len(med_list) > 0 and len(self.med_bag) < 6:
        item = med_list.pop()
        self.med_bag.append(self.objects[item][0])
        self.items_id.remove(item)
        break

			#B9
      if self.step == "pack-medium-items" and len(med_list) > 0 :
        self.get_new_bag()
        break

			#B10
      if self.step == "pack-medium-items" :
        self.step = "pack-small-items"
        print("Packing small items...")
        break

			#B11
      if self.step == "pack-small-items" and len(small_list) > 0 and len(self.small_bag) < 6 and (4 not in self.small_bag):
        item = small_list.pop()
        self.small_bag.append(self.objects[item][0])
        self.items_id.remove(item)
        break

			#B12
      if self.step == "pack-small-items" and len(small_list) > 0 and len(self.small_bag) < 6:
        item = small_list.pop()
        self.small_bag.append(self.objects[item][0])
        self.items_id.remove(item)
        break

			#B13
      if self.step == "pack-small-items" and len(small_list) > 0:
        self.get_new_bag()
        break

			#B14
      if self.step == "pack-small-items":
        self.step="final"
        break 

  def get_new_bag(self):
    if self.step == "pack-large-items":
      self.large_bag_bag.append(self.large_bag)
      self.large_bag = []
    elif self.step == "pack-medium-items":
      self.med_bag_bag.append(self.med_bag)
      self.med_bag = []
    else:
      self.small_bag_bag.append(self.small_bag)
      self.small_bag = []

  def id_to_item(self,ids_list):
    items_name = ""
    for item in ids_list:
      if item in self.objects.keys():
        items_name += self.objects[item][0] + ", "
      else:
        print("\tItem with ID %d not found. Removed from the list."% item)
    return items_name

  def empty_bags(self):
    self.large_bag = []       # Empty large bag
    self.large_bag_bag = []   # Empty large bag of bags
    self.med_bag = []         # Empty med bag
    self.med_bag_bag = []     # Empty med bag of bags
    self.small_bag = []       # Empty small bag
    self.small_bag_bag = []   # Empty small bag of bags
    self.frozen_bag = []      # Empty frozen bag

if __name__ == "__main__":
  Bagger()
```
